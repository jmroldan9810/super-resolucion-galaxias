```python
# Celda 1: Librerías necesarias
import os
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt

# Celda 2: Cargar imagen .FITS
ruta_fits = '../data/raw/dud/n1/n1_0_141190.fits'  # Cambia esto por la ruta real
hdul = fits.open(ruta_fits)
imagen = hdul[1].data

# Celda 5: Guardar como .PNG si quieres trabajar con
```

```python
# Diagnóstico para revisar qué hay en la imagen
print(f"Tipo: {type(imagen)}")
print(f"Dtype: {imagen.dtype}")
print(f"Forma: {imagen.shape if imagen is not None else 'None'}")
print(imagen)
```

```python
hdul.info()  # Te dice qué extensiones existen (0, 1, 2, etc.)
hdul.close()
```

```python

# Celda 3: Visualización de la imagen
plt.figure(figsize=(6,6))
plt.imshow(imagen, cmap='gray', origin='lower')
plt.title('Imagen .FITS')
plt.colorbar()
plt.show()

# Celda 4: Normalización básica (opcional)
imagen_normalizada = (imagen - np.min(imagen)) / (np.max(imagen) - np.min(imagen))

plt.imshow(imagen_normalizada, cmap='gray', origin='lower')
plt.title('Imagen Normalizada')
plt.colorbar()
plt.show()
```
